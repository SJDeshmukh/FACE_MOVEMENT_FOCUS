```python
#########to crop#################FINAL###################################
import os
import cv2
import torch
import time
from facenet_pytorch import MTCNN
# Load the MTCNN face detector
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
mtcnn = MTCNN(keep_all=True, device=device)
cnt=0
# Set minimum and maximum face sizes and confidence threshold
min_face_size = 0
max_face_size = 3000
# frame=cv2.imread("resources/images/iron_man.png")
cam=cv2.VideoCapture(0)
b=None
p_time=0
while True:
    try:
        success,frame=cam.read()
        # Convert the frame to RGB for face detection
        rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        c_time=time.time()
        fps=1/(c_time-p_time)
        p_time=c_time
        print(int(fps))
    #     cv2.putText(rgb,f'FPS: {int(fps)}',(x,y),cv2.FONT_HERSHEY_COMPLEX,1,(255,0,255),3)

        # Detect faces in the RGB frame using the MTCNN detector
        boxes, _ = mtcnn.detect(rgb, landmarks=False)

        # Get the average face size in the current frame
        if boxes is not None:
            face_sizes = [box[2] - box[0] for box in boxes]
            if len(face_sizes) > 0:
                avg_face_size = sum(face_sizes) / len(face_sizes)
                min_face_size = max(30, int(avg_face_size * 0.7))
                max_face_size = int(avg_face_size * 1.3)

        if boxes is not None:
            cnt=1
            a=frame
            for box in boxes:
                x1, y1, x2, y2 = box.astype('int')
                face_size = x2 - x1
                face = frame[y1-50:y2+1000, x1-50:x2+50]
                b=face
                if min_face_size <= face_size <= max_face_size:
                    cv2.rectangle(frame, (x1-50, y1-50), (x2+50, y2+50), (0, 255, 0), 2)
                    w,h,_=face.shape
                    if w>0 and h>0:
                        cv2.imshow("Img",face)
                    else:
                        w,h,_=b.shape
                        frame=cv2.resize(frame,(h,w))
                        cv2.imshow("Img",face)
                else:
                    if cnt>0:
                        w,h,_=b.shape
                        frame=cv2.resize(frame,(h,w))
                        cv2.imshow("Img",frame)
        else:
            if cnt>0:
                width,height,_=b.shape
                frame=cv2.resize(frame,(height,width))
                cv2.imshow("Img",frame)
            else:
                cv2.imshow("Img",frame)
        if cv2.waitKey(1)==27:
            break
        else:
            continue
    except:
        pass

cam.release()
cv2.destroyAllWindows()
# image.upload()
# upload_attendence.upload()
```

```python
import smtplib
import pyshark

def send_email(sender_email, sender_password, recipient_email, subject, message):
    # Set up the SMTP server
    server = smtplib.SMTP('smtp.gmail.com', 587)
    server.starttls()
    server.login(sender_email, sender_password)

    # Create the email message
    email_message = f"Subject: {subject}\n\n{message}"

    # Send the email
    server.sendmail(sender_email, recipient_email, email_message)
    server.quit()

def capture_packets(interface, capture_duration):
    capture = pyshark.LiveCapture(interface=interface, bpf_filter="tcp port 25", output_file="captured_packets.pcap", only_summaries=True)
    capture.sniff(timeout=capture_duration)

    # Print the captured packet information
    for packet in capture:
        print(packet)

# Set the email and network capture parameters
sender_email = "kbtug20192@kbtcoe.org"
sender_password = "f**ko@@fororg#sm"
recipient_email = "kbtug20192@kbtcoe.org"
subject = "Packet Capture"
message = "Please find the attached packet capture file."

interface = "eth0"  # Change to the appropriate network interface
capture_duration = 10  # Capture duration in seconds

# Send the email
send_email(sender_email, sender_password, recipient_email, subject, message)

# Capture packets using Wireshark API
capture_packets(interface, capture_duration)
```
